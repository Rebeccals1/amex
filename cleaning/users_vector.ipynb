```python
US_events_0 = pd.read_csv('https://github.com/ardahk/amex/raw/main/data/US_events_0.csv')
US_events_1 = pd.read_csv('https://github.com/ardahk/amex/raw/main/data/US_events_1.csv')
US_events_2 = pd.read_csv('https://github.com/ardahk/amex/raw/main/data/US_events_2.csv')
US_events_3 = pd.read_csv('https://github.com/ardahk/amex/raw/main/data/US_events_3.csv')
US_events_4 = pd.read_csv('https://github.com/ardahk/amex/raw/main/data/US_events_4.csv')
US_events_5 = pd.read_csv('https://github.com/ardahk/amex/raw/main/data/US_events_5.csv')
```

```python
US_events = pd.concat([US_events_0, US_events_1, US_events_2, US_events_3, US_events_4, US_events_5])
```

```python
users.head()
```

```python
users = users.loc[users['country'] == 'United States']
```

```python
users.shape
```

```python
US_events['user_id'].nunique()
```

```python
purchase_events = US_events.loc[US_events['event_type'] == 'purchase']
```

```python
average_sequence_number = purchase_events.groupby('user_id')['sequence_number'].mean().reset_index()
```

```python
average_sequence_number.columns = ['user_id', 'avg_sequence_number']
```

```python
users_updated = users.merge(average_sequence_number, left_on='id', right_on='user_id', how='left')
```

```python
users_updated.head()
```

```python
users_updated.dropna(inplace = True)
```

```python
users_updated['user_id'].isna().sum()
```

```python
users_updated.drop(columns=['first_name', 'last_name', 'email', 'street_address', 'city', 'latitude', 'longitude', 'created_at', 'country', 'state'], inplace=True)
```

```python
users_updated.head()
```

```python
orders = pd.read_csv('https://github.com/ardahk/amex/raw/main/data/orders.csv')
```

```python
orders.head()
```

```python
orders.drop(columns = ['gender', 'returned_at', 'shipped_at', 'delivered_at'], inplace =True)
```

```python
orders_updated = users_updated.merge(orders, left_on='id', right_on='user_id')
```

```python
orders_updated.head()
```

```python
orders_updated.shape
```

```python
orders_updated.isna().sum()
```

```python
orders_updated.drop(columns = ['user_id_x', 'user_id_y'], inplace = True)
```

```python
orders_updated.head()
```

```python
orders_updated['created_at'] = pd.to_datetime(orders_updated['created_at'], errors='coerce')
orders_updated['month_day'] = orders_updated['created_at'].dt.strftime('%m-%d')
orders_updated.dropna(inplace = True)
orders_updated.drop(columns = ['created_at'], inplace = True)
orders_updated.head()
```
